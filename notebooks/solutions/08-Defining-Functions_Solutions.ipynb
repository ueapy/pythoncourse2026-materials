How to make sections of code reusable and more logical? Create **functions**!

* When you find yourself copying and pasting code in the same program, it is time to create a function.

## Another example

Now for something a little more complex — here we combine a function with a loop to calculate the mean of a list of values.

You might already know that Python has a built-in way to calculate the mean:
```python
import statistics
statistics.mean(temperatures)
```

Or using numpy:
```python
import numpy as np
np.mean(temperatures)
```

We'll use these properly later in the course! For now we are going to build our own mean function from scratch. This is a great way to understand what's happening under the hood, and to practice building functions that take arguments, have defaults, and return values.

`````python
def calc_mean(values):
    total = 0                   #sets the start total to 0
    for value in values:        # keeps going until its run out of values in the list
        total = total + value   # calculates a cumulative total
    mean = total / len(values)  #divides that total by the number of valiues in the list
    
    return mean                 # returns the mean
`````

As mentioned, there are smarter ways to do what we have made above - for example using `sum()` so we don't need a loop at all:
<small>
```python
def calc_mean(values):
    return sum(values) / len(values)
```
</small>

Or just using the built-in mean functions mentioned above. But this example will help you to understand arguments within functions.

Now we have a function named `calc_mean` which takes a single argument `values` (a list of numbers) does something with this argument, and `return`s a value; in this case, the mean of the list:

`````python
temperatures = [12.356, 18.742, 9.431, 22.863, 15.234, 11.698, 19.357]
mean_temperatures = calc_mean(temperatures)

print(mean_temperatures)
`````

* Notice that there is no type information associated with the function inputs or outputs. Python doesn't require you to specify that `values` should be a list, or that the output will be a float. It just trusts you!


* Python functions can return any Python object, a number, a string, a list, or even multiple values at once.

* For example, we could modify our function to return both the mean AND the total:

```python
return mean, total
```

And then unpack both values when we call it:

```python
mean, total = calc_mean(temperatures)
print('Mean:', mean, 'Total:', total)
```

## Adding More Arguments

Consider the `calc_mean` function from before. What if we want to control how many decimal places the result is rounded to? We could try calling it with an extra argument:

`````python
mean_temperatures = calc_mean(temperatures, 1) # one decimal place please!
print(mean_temperatures)
`````

This doesn't work.... why?

We haven't told the function to expect a second argument. Time to redesign it!

`````python
def calc_mean(values, decimal_places):
        # now we have told it to expect how many decimal places we want 
        # with the arument 'decimal_places'

    """
    Calculates the mean of a list of values, 
    rounded to a given number of decimal places
    """
    total = 0
    for value in values:
        total = total + value
    mean = total / len(values)
    
    return round(mean, decimal_places) 
    # AND we have asked it to round the mean to that number
`````

`````python
mean_temperatures = calc_mean(temperatures, 2)

print(mean_temperatures)
`````

### Naming the arguments

You can name the arguments in the function call, in which case the order won't matter.

`````python
# these both work and give the same result
mean_temperatures = calc_mean(decimal_places=1, values=temperatures)
print(mean_temperatures)

mean_temperatures = calc_mean(values=temperatures, decimal_places=1)
print(mean_temperatures)
`````

It is often preferable to name the arguments to:
* make the code easier to read
* avoid mistakes, especially when a function has several arguments and you can't remember the order

## Default Argument Values

Often when defining a function, there are certain values that we want the function to use *most* of the time, but we'd also like to give the user some flexibility. In this case, we can use *default values* for arguments.

For example, we might want `calc_mean` to round to 2 decimal places by default, but still allow the user to change it if they need to:

`````python
def calc_mean(values, decimal_places=2):
    """
    Calculates the mean of a list of values
    
    Parameters
    ----------
    values (list)           : list of numerical values
    decimal_places (int)    : number of decimal places to round to (default=2)
    
    Returns
    -------
    mean (float)    : mean of the input values
    """
    total = 0
    for value in values:
        total = total + value
    mean = total / len(values)
    return round(mean, decimal_places)
`````

`````python
# uses the default - 2 decimal places
mean_temperatures = calc_mean(temperatures)
print(mean_temperatures)
`````

`````python
# override the default
mean_temperatures = calc_mean(temperatures, decimal_places=4)
print(mean_temperatures)
`````

Notice that our function now has two types of arguments:

* `values` - must always be provided, and position matters
* `decimal_places` - has a default value of 2, so it's optional

We'll come back to what `*args` and `**kwargs` actually mean shortly!

`````python
# This won't work - why?
mean_temperatures = calc_mean(4, temperatures)
print(mean_temperatures)
`````

This fails because `4` is passed as the first positional argument. So Python thinks `4` is the list of values and `temperatures` is the number of decimal places. Python can't iterate over 1 number so it crashes.

This is why naming your arguments matters:

<small>

```python
# This works - we are explicit about which argument is which even though they are in different places
mean_temperatures = calc_mean(decimal_places=4, values=temperatures)
print(mean_temperatures)
```

```python
# This also works - correct order, this is what python is expecting even though we have not used a keyword
mean_temperatures = calc_mean(temperatures, 4)
print(mean_temperatures)
```

</small>

A good rule of thumb, if your function has more than two arguments, name them when you call it. It makes your code much easier to read and avoids this kind of mistake both by python and by you!

## `*args` and `**kwargs`: flexible arguments

Sometimes you want a function that can accept **any number** of arguments without defining them all in advance.

* `*args` collects any number of **unnamed** positional arguments
* `**kwargs` collects any number of **named** keyword arguments

You'll see this a lot in libraries like matplotlib — `plt.plot()` can accept any number of styling options (`color`, `linewidth`, `label` etc.) because it uses `**kwargs`.

Here's a simple example:


`````python
# *args example - pass in as many numbers as you like
def add_numbers(*args):
    return sum(args)

print(add_numbers(1, 2, 3)) 
print(add_numbers(10, 20))    
`````

`````python
# **kwargs example - pass in any named options
def describe_pet(**kwargs):
    for key, value in kwargs.items():
        print(key, ":", value)

describe_pet(name="Clifford", species="Dog", age=5, colour='red')
`````

So when we call `calc_mean(temperatures, decimal_places=2)`:

* `*args` receives `(temperatures,)` - the unnamed positional value
* `**kwargs` receives `{'decimal_places': 2}` - the named keyword value

It's not the names `args` and `kwargs` that matter - it's the `*` and `**` characters. You could call them anything - `*inputs` and `**options` would work just as well.

They are just a storage mechanism, much like a variable!

You won't need to write functions like this often, but you'll see them regularly in packages you use.

# The bones of a function

When you are making your own function there is a good practice convention to use. This is useful if you are exchanging functions with others so they understand how your function works. Or at the least your future self knows what past you was doing!

The bare bones of how a function should be laid out is below:

`````python
def my_function_name(parameter1, parameter2=0.1):    
    ''' #note: triple quotes let us write strings that span multiple lines!
    
    A description of what the function does.
    
    
    Parameters - a list of the possible input parameters, their type and what they are
    ----------
    parameter1 (type e.g. int)    : Description of what parameter 1 should be
 
    parameter2 (type e.g. float)  : Description of what parameter 2 should be

    Returns
    -------
    output    : A description of the output of the function
    
    '''
    
    #This is where the code goes!
    output = parameter1+parameter2
    
    return output  #and here is where you specify what it outputs
`````

`````python
def print_promise(my_name):
    '''
    
    This code takes a name and uses it to print a promise!
    # Tip: it is always a good idea to name your function as the action it does!
    
    Parameters 
    ----------
    my_name (type e.g. str)    : A string
 

    Returns
    -------
    promise    : A string including the parameter my_name
    
    '''
    
    promise = ('I, {name}, will always comment my code!'.format(name = my_name))
    
    return promise
`````

`````python
def calc_co2e(dist,
              returnf=False,
              firstclass=False,
              radforc=2.0):
    """
    calculate equivalent carbon emissions from flights
    
    Parameters
    ==========
    dist - flight distance in km
    
    Optional inputs
    ---------------
    returnf - Return flight (default=False)
    firstclass - First class flight (default=False)
    radforc - radiative forcing factor (default=2.0)
    
    Returns
    =======
    CO2 equivalent emissions in kg calculated as
    co2e = dist * emission factor

    Emission factors (kg CO2e/pkm)
    https://flygrn.com/blog/carbon-emission-factors-used-by-flygrn
    
    0.26744  < 700 km 
    0.15845  700 – 2500 km
    0.15119  > 2500 km 
    
    First Class and return flights both double the carbon emissions.
    """
    
    # Select emission factor based on distance
    if dist < 700:
        ef = 0.26744
    elif dist <= 2500:
        ef = 0.15845
    else:
        ef = 0.15119
    
    # Base emissions
    co2e = dist * ef
    
    # Double emissions if return flight
    if returnf:
        co2e = co2e * 2
    
    # Double emissions if first class
    if firstclass:
        co2e = co2e * 2

    
    # Apply radiative forcing factor
    co2e = co2e * radforc
    
    return co2e
`````

`````python
distance = 1600
co2_emissions = calc_co2e(distance, returnf=True)
print(f"CO2e emissions for London-Sevilla return flight: {co2_emissions:.2f} kg")
`````

## Bonus Material: Anonymous (``lambda``) Functions

``"lambda" [parameter_list]: expression``

Earlier we covered the most common way of defining functions, the ``def`` statement.
There is another way of defining a short, one-off function with the ``lambda`` statement.
It looks something like this:

`````python
add = lambda x, y: x + y
add(1, 2)
`````

This lambda function is equivalent to

`````python
def add(x, y):
    return x + y
`````

So why would you want to use lambda functions?
Primarily, it comes down to the fact that *everything is an object* in Python, even functions themselves!
That means that functions can be passed as arguments to functions.

As an example of this, suppose we have some data stored in a list of dictionaries:

`````python
data = [{'first':'Guido', 'last':'Van Rossum', 'YOB':1956},
        {'first':'Grace', 'last':'Hopper',     'YOB':1906},
        {'first':'Alan',  'last':'Turing',     'YOB':1912}]
`````

Now suppose we want to sort this data.
Python has a ``sorted`` function that does this:

`````python
sorted([2,4,3,5,1,6])
`````

We need a way to tell the function *how* to sort our data.
We can do this by specifying the ``key`` function, a function which given an item returns the sorting key for that item:

`````python
# sort alphabetically by first name
sorted(data, key=lambda item: item['first'])
`````

`````python
# sort by year of birth
sorted(data, key=lambda item: item['YOB'])
`````

While these key functions could certainly be created by the normal, ``def`` syntax, the ``lambda`` syntax is convenient for such short one-off functions like these.

This is how it would look like using a standard function.

`````python
def by_lastn(dic):
    return dic['last']

sorted(data, key=by_lastn)
`````
