### Sea level rise

Given the current rate of sea level rise, calculate the predicted sea level in the future.

* `current_sea_level = 0.2` metres above 1980 baseline
* `rise_per_year = 0.0033` metres per year (current average rate)
* `years = 50`

Calculate:
- The predicted sea level in 50 years time relative to the 1980 baseline
- The total rise over that period

Print both results using f-strings.

````python
current_sea_level = 0.2
rise_per_year = 0.0033
years = 50

predicted_level = current_sea_level + (rise_per_year * years)
total_rise = rise_per_year * years

print(f"Predicted sea level in {years} years: {predicted_level:.4f} metres")
print(f"Total rise: {total_rise:.4f} metres")
````

#### Exercise 2:
* Use the list of country names below
* Create a new list where each country name is capitalised correctly

Hint: use the string method `.capitalize()`

````python
# Exercise 2 - simple solution
countries = ['paRaguay', 'uRUGUAY', 'SPAIN', 'chile', 'el Salvador']
print([country.title() for country in countries])

# Note: capitalize() won't handle 'el Salvador' correctly
# For multi-word country names use .title() or the longer solution below:
newnames = []
for country in countries:
    if len(country.split()) > 1:
        newname = [word.capitalize() for word in country.split()]
        newname = ' '.join(newname)
    else:
        newname = country.capitalize()
    newnames.append(newname)
print(newnames)
````

## 7.1 Putting it all together — Field Survey Summary

You've collected data from 5 survey stations. Using everything you've learned today, build a mini survey summary.

```python
stations = ['NorthSea_01', 'NorthSea_02', 'NorthSea_03', 'NorthSea_04', 'NorthSea_05']
temperatures = [8.2, 7.9, 9.1, 6.8, 8.5]   # celsius
salinities = [34.2, 35.1, 33.8, 35.6, 34.9] # PSU
```

**Tasks:**
1. Create a new list of station names in uppercase
2. Calculate the mean temperature (hint: use `sum()` and `len()`)
3. Create a list of temperatures that are above the mean
4. Build a dictionary for the first station containing its name, temperature and salinity
5. Print a formatted summary using f-strings, e.g:
   * Number of stations
   * Mean temperature
   * How many stations are

````python
stations = ['NorthSea_01', 'NorthSea_02', 'NorthSea_03', 'NorthSea_04', 'NorthSea_05']
temperatures = [8.2, 7.9, 9.1, 6.8, 8.5]
salinities = [34.2, 35.1, 33.8, 35.6, 34.9]

# 1. Station names in uppercase
stations_upper = [s.upper() for s in stations]
print(stations_upper)

# 2. Mean temperature
mean_temp = sum(temperatures) / len(temperatures)
print(f"Mean temperature: {mean_temp:.2f}°C")

# 3. Temperatures above the mean
above_mean = [t for t in temperatures if t > mean_temp]
print(f"Temperatures above mean: {above_mean}")

# 4. Dictionary for first station
first_station = {
    'name': stations[0],
    'temperature': temperatures[0],
    'salinity': salinities[0]
}
print(first_station)

# 5. Formatted summary
print(f"\n--- Survey Summary ---")
print(f"Number of stations: {len(stations)}")
print(f"Mean temperature: {mean_temp:.2f}°C")
print(f"Stations above mean temperature: {len(above_mean)} out of {len(stations)}")
````

````python
# Example: finding the station with the highest temperature
# This uses max() with a key argument — something we haven't covered yet
# but is a good example of what AI might suggest

max_temp = max(temperatures)
max_index = temperatures.index(max_temp)
max_station = stations[max_index]

print(f"Highest temperature: {max_temp}°C at {max_station}")

# Example: list of dictionaries — one per station
survey_data = [{'name': s, 'temperature': t, 'salinity': sal} 
               for s, t, sal in zip(stations, temperatures, salinities)]

for entry in survey_data:
    print(entry)
````
