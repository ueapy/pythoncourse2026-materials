# Use of AI to help you code 

Programming AIs (large language models, or LLMs)  can save you time by generating code for you. They can also be used to help you learn to program. However, it is important to use AI wisely. It is not a substitute for learning how to code. And it doesn't know  what you want unless you tell it.

## General-purpose generative AI vs. coding AI
General-purpose AI can be prompted to generate programming code. This is done outside your programming environment and makes it difficult to integrate the suggestions with your existing code. Code-optimised AI can be accessed from with an Integrated Development Environment (IDE) like VSCode, PyCharm, Spyder or Jupyter lab. There are several advantages of using AI this way:
- It is aware of the rest of your script or notebook. It can also be aware of the code directory (e.g. open directory in VSCode).
- Suggests code as you write
- Autogenerates documentation (docstrings and comments)
- Prompts can be written in the code or prompt window
- Can run code and debug it (it you enable agents)

##  Principles of how to use AI


### 1. Break down the tasks.
   - Ask for functions with clear inputs and outputs so that it is easier to test, modify and integrate with your code.
   - Asking for code blocks is trickier as it requires reuse of variables and can have unexpected knock on effects on the code.
   - Use code suggestions to ask for simpler commands without having to lookup the documentation.

### 2. Write good prompts.
   - Specify everything.
       - What is the input, output and logic?
       - Understand that what is not specified will be made up.
   - Iterate on prompts.
       - If the results is wrong or incomplete keep adding specifications until you get the desired result.
   - Specify any packages or methods you want to use.
       - For example, if you want to use `numpy` to generate random numbers, specify it in the prompt. Otherwise, the AI might use the built-in `random` module or any other method to generate random numbers.

### 3. Ask for explanations when needed.

  - Explain existing code
  - Further explain suggested code

### 4. Treat the AI as an assistant, not as the creator.

   - Review everything.
   - Verify the logic.
   - Test the code.

AI generated code tends to have more bugs than human generated code. AI models often produce "shallow" code that is syntactically correct but functionally flawed.
